This program approaches f(x)=cos(x) function in between 0<x<pi/2 interval with Taylor Expansion. It uses [SciPy.org](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.approximate_taylor_polynomial.html) library to create Taylor Polynomial.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial
```

```python
exp_error=input('Please enter the error rate ->')
exp_error=np.float(exp_error)
```

```python
x=np.linspace(0,np.pi/2,num=100)
```

```python
for degree in np.arange(1,15,step=1):
    cos_taylor=approximate_taylor_polynomial(np.cos,0,degree,1,order=None)
    cal_error=approximate_taylor_polynomial(np.cos,0,degree+1,1,order=None)-approximate_taylor_polynomial(np.cos,0,degree,1,order=None)
    #Epsilon number is chosen pi/2 to calculate the maximum error
    if abs(cal_error(np.pi/2))<exp_error:
        break
```

```python
print(abs(cal_error(np.pi/2)))
```

```python
print(cos_taylor(0))
```

```python
plt.plot(x,np.cos(x),color='yellow',marker='*',label='cos curve')
plt.plot(x,cos_taylor(x),color='red',label=f'degree={degree}')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.0, shadow=True)
plt.tight_layout()
plt.axis([-0.1, 1.75, -0.25, 1.1])
plt.show()
```
